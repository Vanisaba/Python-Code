```python
import numpy as np  
import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches
from sklearn import decomposition  
from sklearn import datasets 
import pandas as pd
```

```python
#prepare the dataframe
iris_array=datasets.load_iris()
iris_array
iris_df=pd.DataFrame(iris_array.data,columns=iris_array.feature_names)

iris_df['target']=iris_array.target
iris_df['species']=iris_df['target'].map({0:iris_array.target_names[0],1:iris_array.target_names[1],2:iris_array.target_names[2]})
#iris_df.drop(['target'],axis=1,inplace=True)
iris_df
```

```python

fig=plt.figure(figsize=(20,20))
ax = fig.add_subplot(1, 1, 1 )
ax = fig.gca(projection='3d')

col=iris_df['target']
x=iris_df['sepal length (cm)']
y=iris_df['sepal width (cm)']
z=iris_df['petal length (cm)']
plt.style.context('dark_background')
 
colors = {'setosa':'red', 'versicolor':'blue','virginica':'green' }
labels={0:'setosa',1:'versicolor',2:'virginica'}

ax.scatter(x ,y,z ,c=iris_df['species'].apply(lambda x:colors[x])  ,cmap=plt.cm.Set1, edgecolor='k', s=80  )
ax.set_title(' Iris Dataset')
ax.set_xlabel('SEPAL LENGTH')
ax.set_ylabel('SEPAL WIDTH')
ax.set_zlabel('PETAL LENGTH')
r_patch=mpatches.Patch(color='red', label='setosa'  )
b_patch=mpatches.Patch(color='blue', label='versicolor' )
g_patch=mpatches.Patch(color='green', label='virginica' )
ax.legend(handles=[r_patch,b_patch,g_patch])

plt.show()
```
